```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def y_function(x):
    return x**2

def y_derivative(x):
    return 2*x
```

```python
x=np.arange(-100, 100, 0.1)
y=y_function(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
current_pos=(80,y_function(80))
```

```python
plt.plot(x,y)
plt.scatter(80,y_function(80), color='red')
plt.show()
```

```python
learning_rate=0.01
```

```python
for k in range(1000):
    new_x=current_pos[0]-learning_rate*y_derivative(current_pos[0])
    new_y=y_function(new_x)
    current_pos=[new_x,new_y]
```

```python
plt.plot(x,y)
plt.scatter(80,y_function(80), color='red')
plt.pause(0.01)
plt.clf()

```
